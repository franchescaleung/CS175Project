Jupyter notebook to parse and create an edited collection of dwight-based queries and responses

```python
from schrutepy import schrutepy
```

```python
df = schrutepy.load_schrute() #load the data into a dataframe
```

```python
f = open("dwight_text.txt", 'w') #open file to write to
listens = [] #make a list of queries and replies
replies = []
vocab = {} #make a dictionary to hold words seen
```

```python
for i in range(len(df.character)): #for each entry in the data frame
    if df.character[i] == "Dwight": #check to see if the chracter speaking is dwight
        listen =  str(df.text[i-1]) #get the previous utterance in the script 
        if len(listen.split(' ')) > 6: #if utterance is greater than 6 words we skip it -> too verbose
            continue 
        print(listen) #print it to screen to make sure its doing something
        listen = listen.replace('"', '') #remove " characters
        reply = str(df.text[i]) #get utterance at position i in the script
        reply = reply.replace('"', '') #remove " characters
        if len(reply.split(' ')) > 6: # if reply is greater than 6 words we skip it ->too verbose
            continue
        for word in listen.split(' '): #add each word in the query to the dictionary
            if word in vocab:
                vocab[word] += 1
            else:
                vocab[word] = 1
        for word in reply.split(' '): #add each word in the response to the dictionary
            if word in vocab:
                vocab[word] += 1
            else:
                vocab[word] = 1
        listens.append(listen) #add query and response to two seperate lists
        replies.append(reply)      
```

```python
print(len(listens))
for i in range(len(listens)): #for each query in our list
    flag = False #initialize boolean value
    print("entry: ", listens[i]) #print it for debugging
    for word in listens[i].split(' '): #break the query into seperate words
        print(vocab[word]) #debug
        if vocab[word] <= 3: #if any of the words occurs less than 4 times in the whole script set it to true
            flag = True
    if flag == True: #if we flipped the bool remove both the query and response from the lists
        listens.pop(i)
        replies.pop(i)
```

```python
print(len(listens)) 
```

```python
for i in range(len(listens)): #write each query response pair to our outfile
    f.write('{"listen": "' + listens[i] + '", "reply": "' + replies[i] + '"}\n') 
```

```python
f.close() #close outfile
```
